```python
#tips_df[tips_df['day']=='Sun']['day'].count()
#C Style
days=tips_df.day.unique()

day_data={}
for each_day in days:
 day_data[each_day]=[tips_df[tips_df['day']==each_day]['day'].count()]

print(day_data)
```

```python
day_df=pd.DataFrame(day_data)
sns.barplot(day_df)
```

```python
tips_df.groupby('day').count() # Python style
```

```python
tips_df.groupby('day')['day'].count() # Python style
```

```python
tips_series=tips_df.groupby('day')['day'].count() # Python style
sns.barplot(x=tips_series.index,y=tips_series.values)
```

```python
tips_series=tips_df.groupby('sex')['day'].count() # Python style
sns.barplot(x=tips_series.index,y=tips_series.values)
```
